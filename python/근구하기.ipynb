#### 근 구하기
- 원하는 근을 중심으로 서로 다른 부호를 갖는 구간을 찾는다.
- 구하고자 하는 근의 해상도가 주어지면 10배정도의 resolution으로 함수를 근사화한다.
- 함수의 절대 값이 가장 작은 x를 찾으면 됨.

```python
# Roots of an equation
import numpy as np
from scipy.optimize import root
import matplotlib.pyplot as plt
x=np.linspace(-np.pi,np.pi,100)
y=x+np.cos(x)
plt.plot(x,y)
plt.grid()
```

```python
res=1
h=10**(-res)
x=np.arange(-1,0,h)
y=x+np.cos(x)
plt.plot(x,y,'o')
plt.grid()
plt.figure()
plt.plot(x,np.abs(y),'o')
plt.grid()
```
